```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ingest data

df = pd.read_csv(r"C:\Users\asunday\Downloads\london_weather.csv")
df.head()
```

```python
df.info()
```

```python
df['date'] = df['date'].astype('str')
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df.head()
```

```python
df.set_index('date', inplace = True)
df.head()
```

```python
# resampling the data into yearly data

annualData = df.resample('Y').mean()
annualData.head()
```

```python
x = annualData.index
y = annualData['sunshine']
plt.figure(
    figsize = (20, 7)
)
plt.plot(x, y)
plt.show()
```
