#### Creating Basic Plots in Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\asunday\Downloads\calories.csv")
df.head(2)
```

```python
df.dropna(inplace = True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# Creating a line plot
plt.figure(
    figsize=(20, 7)
)
plt.plot(
    df['Date'], 
    df['Pulse'], 
    linestyle='--',
    color='red',
    marker='o',
    label = 'Pulse',
    linewidth=1
)
plt.title('Daily Pulse Trend')
plt.xlabel('Date')
plt.ylabel('Pulse')
plt.legend()
plt.show()
```

```python
# Creating a barplot

plt.figure(
    figsize=(20, 7)
)
plt.bar(
    df['Date'], 
    df['Maxpulse'],
    color='g'
)
plt.title('Daily Max Pulse Trend')
plt.xlabel('Date')
plt.ylabel('MaxPulse')
plt.show()
```

```python
# Creating a scatter plot

plt.figure(
    figsize=(14, 7)
)
plt.scatter(
    df['Calories'], 
    df['Pulse'],
)
plt.title('Scatter plot of Calories vs. Pulse')
plt.xlabel('Calories')
plt.ylabel('Pulse')
plt.show()
```
