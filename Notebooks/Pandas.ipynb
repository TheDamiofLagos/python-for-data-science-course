##### Merge, Join, and Concatenate

```python
# Merge

# Merging allows us to combine two or more DataFrames based on common columns or indices.

# Syntax: pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None)
```

```python
df1 = pd.read_csv(r"C:\Users\asunday\Downloads\LOTR.csv")
df2 = pd.read_csv(r"C:\Users\asunday\Downloads\LOTR 2.csv")

df1
```

```python
df2
```

```python
df1.merge(df2, how = 'left', on = ['FellowshipID', 'FirstName'])
```

```python
# Join

# Definition: Join is a convenient method for combining DataFrames using their indices.
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix= 'l', rsuffix = 'r')
```

```python
df1_b = df1.set_index('FellowshipID')
df1_b
```

```python
df2_b = df2.set_index('FellowshipID')
df2_b
```

```python
df1_b.join(df2_b, how = 'inner', lsuffix= 'l', rsuffix = 'r')
```

```python
# Concatenate 

df1.append(df2)
```
