```python
# import libraries

import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\asunday\Downloads\world_population.csv")
df.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x) 
# displays my floats in 2decimal points
```

```python
df2 = df.groupby('Continent')[[
    '1970 Population', 
    '1980 Population', 
    '1990 Population',
    '2000 Population',
    '2010 Population',
    '2015 Population',
    '2020 Population',
    '2022 Population'
]].mean().sort_values(by = '2022 Population', ascending = False)

df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot(
    figsize = (20, 7),
    title = 'Average population for each continent per census year'
);
```
